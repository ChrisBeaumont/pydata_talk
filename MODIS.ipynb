```python
%matplotlib inline
from scidbpy import connect

import matplotlib.pyplot as plt
plt.rcParams['figure.figsize']=(12, 8)
plt.rcParams['font.size'] = 1
```

```python
sdb = connect()
elev = sdb.wrap_array('all')
elev
```

## An irregular grid of elevation samples. Here's a small subset

```python
pts = elev.sdbslice[365000:365500, -1125500:-1125000].tosparse()

plt.scatter(pts['lon_e4'], pts['lat_e4'], s=100, c=pts['altitude'], cmap='Greens_r')


plt.xlim(0, 500)
plt.ylim(0, 500)
plt.xticks([])
plt.yticks([])
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.title('Elevation Samples')
```

## Use SciDB to aggregate the data onto a regular grid

```python
# Example plots of MODIS elevation data only. The example subset of MODIS used here consists of only 3 tiles and is about 4GB.

f = sdb.afl  # bindings to SciDB Array Functional Language

def make_map(lat, lon, extent, image_height=200):

    array_scale = 1e4    
    def toindex(x):
        return int(x * array_scale)

    # compute pixel coordinates of subarray
    lon_start = toindex(lon - extent)
    lon_end   = toindex(lon + extent)
    lat_start = toindex(lat - extent)
    lat_end   = toindex(lat + extent)

    regrid_factor = int(2 * extent * array_scale / image_height)
    
    # extract subarray
    im = elev.sdbslice[lat_start:lat_end, lon_start:lon_end]
    
    # rebin to requested image size
    im = im.regrid(regrid_factor)
    
    # download into numpy array
    im = im.toarray()
 
    # visualize
    plt.imshow(im, origin='lower', cmap='Greens_r', 
               extent=[lon - extent, lon + extent, lat - extent, lat + extent])
    plt.colorbar()
    plt.xlabel("Longitude")
    plt.ylabel("Latittude")
    plt.title("Alititude")
```

```python
make_map(lat=34, lon=-113, extent=1.5)
```
