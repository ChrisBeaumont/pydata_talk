# Why is the Sky Blue? Why is the Sunset Red?

* Light is absorbed and scattered when it passes through material

* Blue light is affected more

* At sunset, the sun skims through lots of atmosphere, attenuating all the blue light.

By mesuring colors at several wavelengths, we can figure out how much material lies between us and a stars. We can use this to make **density maps** of interstellar clouds.


### Here's the math

```python
%%latex 

\begin{eqnarray}
 C =& \begin{pmatrix}
0.028 + \sigma_J^2 + \sigma_H^2 &  0.00293 - \sigma_H^2 \\
0.00293 - \sigma_H^2 &  0.02765 + \sigma_H^2 + \sigma_K^2 \\
\end{pmatrix} \\
 k   =& [0.106,~ 0.063] \\
 w   =& \frac{C^{-1}k}{k C^{-1}k} \\
 \bf{A_v} =& w \cdot [J - H - 0.487, H - K - 0.134] \\
\end{eqnarray}
```

# Implemented with SciDB operators

```python
def nicer(array):
    # Build C Matrix
    q = afl.apply(array, 
                'CA', '0.028 + pow(jsig, 2) + pow(hsig, 2)',
                'CB', '0.00293 - pow(hsig, 2)',
                'CD', '0.02765 + pow(hsig, 2) + pow(ksig, 2)')
    # 1/det(C)
    q = afl.apply(q, 'detCinv', '1/(CA * CD - CB * CB)')

    # C^-1 k
    q = afl.apply(q,
                'CK1', 'detCinv * CD * 0.106 - detCinv * CB * 0.063',
                'CK2', '-detCinv * CB * .106 + detCinv * CA * 0.063')

    # W
    q = afl.apply(q,
                'W1', 'CK1 / (0.106 * CK1 + 0.063 * CK2)',
                'W2', 'CK2 / (0.106 * CK1 + 0.063 * CK2)')

    #Av
    avs = afl.papply(q, 'Av', 'W1 * (jmh - 0.487) + W2 * (hmk - 0.134)')
        
    return avs
```

```python
def nicer_map(lon, lat, extent=5, image_height=300):
    extent = 5
    array_scale = 10000   # pixels / deg

    def idx(x):
        return int(x * array_scale)

    lat_start = idx(lat - extent)
    lat_end   = idx(lat + extent)
    lon_start = idx(lon - extent)
    lon_end   = idx(lon + extent)
    factor = int(2.0 * extent / image_height * array_scale)

    
    # Here's the SciDB stuff
    array = sdb.wrap_array('twomass_2d')  #grab array
    query = nicer(array)   # Derive reddening
    query = afl.subarray(query, lon_start, lat_start, lon_end, lat_end)  # look at a sub-window
    query = afl.regrid(query, factor, factor, 'avg(Av)')  # average nearby points

    #execute query, convert to NumPy array
    im = query.eval().substitute(0).toarray()

    #visualize
    fig, ax = plt.subplots()
    im = ax.imshow(im.T, origin='lower', cmap='bone_r', 
                   extent=[lon - extent, lon + extent, lat - extent, lat + extent], 
                   vmin=0.5, vmax=15)

    ax.set_xlim(lon + extent, lon - extent)
    fig.colorbar(im, ax=ax)
    return fig
```

```python
oph = nicer_map(247., -24, image_height=200)
```

# Up the ladder of abstraction...

```python
from ipywidgets import StaticInteract, RangeWidget, RadioWidget

def show(region, resolution):
    locs = {'rhoOph':[247., -24],
            'Pipe': [260., -25.5],
            'Taurus': [65., 27.]}
    lon, lat = locs[region]
    return nicer_map(lon, lat, image_height=resolution)

StaticInteract(show, 
               resolution=RangeWidget(100, 400, 100), 
               region=RadioWidget(['rhoOph', 'Pipe', 'Taurus']))
```
